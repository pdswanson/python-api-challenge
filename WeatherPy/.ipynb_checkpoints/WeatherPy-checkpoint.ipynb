# WeatherPy

### Observations and Insights

* The temperature range is relatively flat (i.e., no noticeable peak at the equator) for cities between the 20th parallel north and the 20th parallel south, and only begins to drop outside of those latitudes in either direction.
* The driest cities (humidities of 40% or lower) in the Northern Hemisphere are all located between the equator and the 50th parallel, but virtually all of the driest cities in the Southern Hemisphere can be found south of the 20th parallel.
* Cities in the Northern Hemisphere tend to be cloudier the farther they're located from the equator, while cities in the Southern Hemisphere tend to be cloudier the closer they are to the equator.
* Wind speeds for cities in both hemispheres are mildly correlated, gradually increasing based on the proximity to their respective poles.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress
from api_keys import weather_api_key
from citipy import citipy

output_data_file = "output_data/cities.csv"

lat_range = (-90, 90)
lng_range = (-180, 180)
```

### Generate Cities List

```python
lat_lngs = []
cities = []

lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    if city not in cities:
        cities.append(city)

len(cities)
```

### Perform API Calls

```python
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&"
query_url = url + "APPID=" + weather_api_key + "&q="
```

```python
city_name = []
cloudiness = []
country = []
date = []
humidity = []
latitude = []
longitude = []
max_temp = []
wind_speed = []

count = 1
sets = 1
```

```python
print(f"Beginning Data Retrieval")
print(f"-------------------------------")

for city in cities:
    try:
        response = requests.get(f"{query_url}{city}").json()
        city_name.append(response["name"])
        cloudiness.append(response["clouds"]["all"])
        country.append(response["sys"]["country"])
        date.append(response["dt"])
        humidity.append(response["main"]["humidity"])
        latitude.append(response["coord"]["lat"])
        longitude.append(response["coord"]["lon"])
        max_temp.append(response["main"]["temp_max"])
        wind_speed.append(response["wind"]["speed"])
        print(f'Processing Record {count} of Set {sets} | {city}' )
        count += 1

    except:
        print("City not found. Skipping...")
        continue
         
    if count == 51:
        count = 1
        sets += 1
        time.sleep(1)

print(f"-------------------------------")
print(f"Data Retrieval Complete")
print(f"-------------------------------")
```

### Convert Raw Data to DataFrame

```python
city_info = {
    'City': city_name,
    'Cloudiness': cloudiness, 
    'Country': country,
    'Date': date, 
    'Humidity': humidity,
    'Latitude': latitude, 
    'Longitude': longitude, 
    'Max Temp': max_temp,
    'Wind Speed': wind_speed
}

city_data = pd.DataFrame(city_info)
city_data.count()
```

```python
city_data.to_csv('Resources/WeatherPy.csv', encoding='utf-8', index=False)
city_data.head()
```

## Linear Regression

```python
no_hemis = city_data.loc[city_data['Latitude'] >= 0.0]
no_hemis.reset_index(inplace=True)

so_hemis = city_data.loc[city_data['Latitude'] < 0.0]
so_hemis.reset_index(inplace=True)
```

####  Northern Hemisphere - Max Temp (F) vs. Latitude Linear Regression

```python
x_values = no_hemis['Latitude']
y_values = no_hemis['Max Temp']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
print(f"The r-squared is: {rvalue}")
```

```python
plt.scatter(x_values, y_values, marker="o", facecolors="blue", edgecolors="black")
plt.plot(x_values, regress_values, "r-")
plt.annotate(line_eq, (5,-10), fontsize=15, color="red")
plt.title('City Latitude vs. Max Temps: Northern Hemisphere')
plt.xlabel('Latitude')
plt.ylabel('Max Temperature (F)')
plt.grid(linestyle='dashed')
plt.savefig("Images/no_lat_vs_temp.png")
plt.show()
```
